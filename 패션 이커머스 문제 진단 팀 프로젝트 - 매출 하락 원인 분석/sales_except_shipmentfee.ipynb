```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import os
```

```python
# transaction table 가져오기
base_path = os.getcwd()
transaction_new = pd.read_csv(base_path + '/transaction_final.csv')
transaction_new.info()
```

```python
transaction_new.sort_values(by=['customer_id', 'created_at'], inplace=True)
transaction_new.head()
```

```python
# customer_id, booking_id 별 total_amount와 shipment_fee 차이
sales = pd.DataFrame(transaction_new.groupby(['customer_id', 'booking_id']).total_amount.unique() - transaction_new.groupby(['customer_id', 'booking_id']).shipment_fee.unique())
sales
```

```python
# 데이터프레임 재구조화
sales = sales.reset_index()

# 컬럼명 변경 (0->sales)
sales.rename(columns={0:'sales'}, inplace=True)

# sales 컬럼 형식 ([] -> int)
sales.sales = sales.sales.apply(lambda x:x[0])
sales
```

```python
sales.to_csv('sales.csv', index = False)
```
